**This is dedicated to show how the required csv file for the "Datascience Methodology using Cuising Prediction" is prepared from the json files available.**

```python
#importing required libraries
import pandas as pd
import numpy as np
import json
```

## Train_Test_Data from train.json

```python
with open('train.json') as data_file:    
    traindata = json.load(data_file)
```

```python
def extract_json(filename):
    ids = []
    cuisines = []
    ingredients = []    
    
    for each in filename:
        ids.append(each['id'])
        cuisines.append(each['cuisine'])
        ingredients.append(each['ingredients'])
        
    d = {'id':ids, 'cuisine':cuisines, 'ingredients': ingredients} 
    dataframe = pd.DataFrame(d, columns=['id','cuisine','ingredients'])
    return dataframe
```

```python
dataset = extract_json(traindata)
```

```python
dataset
```

```python
def do_one_hot(dataframe):
    df_new = pd.DataFrame()
    #For every row in the dataframe, iterate through the list of genres and place a 1 into the corresponding column
    for index, row in dataframe.iterrows():
        for each in row['ingredients']:
            df_new.at[index, each] = 1
        print(index)
    #Filling in the NaN values with 0 to show that a movie doesn't have that column's genre
    df_new = df_new.fillna(0)
    dataframe = dataframe.drop(['ingredients'], axis=1)
    dataframe = pd.concat([dataframe, df_new], axis=1)
    return dataframe
```

```python
dataset_final = do_one_hot(dataset)
```

```python
dataset_final
```

```python
dataset_final.to_csv('train_validation_data.csv', index=False)
```
